```python
import random
#comments show where the code was changed

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 10)
        print(' ')

def check_winner(board, mark):
    win_conditions = [
        [board[0][0], board[0][1], board[0][2]],
        [board[1][0], board[1][1], board[1][2]],
        [board[2][0], board[2][1], board[2][2]],
        [board[0][0], board[1][0], board[2][0]],
        [board[0][1], board[1][1], board[2][1]],
        [board[0][2], board[1][2], board[2][2]],
        [board[0][0], board[1][1], board[2][2]],
        [board[2][0], board[1][1], board[0][2]],
    ]
    return [mark, mark, mark] in win_conditions

def get_empty_positions(board):
    positions = []
    for i in range(3):
        for j in range(3):
            if board[i][j] == " ":
                positions.append((i, j))
    return positions

def user_move(board):
    while True:
        try:
            row = int(input("Enter the row (1-3): ")) - 1
            col = int(input("Enter the column (1-3): ")) - 1
            if board[row][col] == " ":
                board[row][col] = "X"
                break
            else:
                print("This position is already taken.")
        except (ValueError, IndexError):
            print("Invalid input. Please enter numbers between 1 and 3.")

def artificial_intelligence(board, depth, is_maximizing): #added a minimax type function that is able to see which move would be the best for the computer
    if check_winner(board, "O"):      #calls the check_winner method to see if there is any winining moves for the mark O
        return 1
    elif check_winner(board, "X"):    #calls the check_winner method to see if there is any winining moves for the mark X
        return -1
    elif not get_empty_positions(board):  #returns 0 if board is not empty and no moves seen, aka draw
        return 0

    if is_maximizing:   #maximizing the score for the 'O' player
        best_score = -float('inf')
        for (i, j) in get_empty_positions(board):
            board[i][j] = "O"         #places a 'O' in all empty positions
            score = artificial_intelligence(board, depth + 1, False)    #recursively calls the AI method, increasing the depth
            board[i][j] = " "   #removes the mark
            best_score = max(score, best_score) #sees which move has given the best score so far
        return best_score #returns that score
    else:
        best_score = float('inf')     #minimizing the score for the 'X' player, same things like above, but now minimizing the score
        for (i, j) in get_empty_positions(board):
            board[i][j] = "X"
            score = artificial_intelligence(board, depth + 1, True) #checks the next AI comptuer move
            board[i][j] = " "
            best_score = min(score, best_score)
        return best_score

def computer_move(board):
    best_score = -float('inf')  #starts best score at -infinity
    best_move = None
    for (i, j) in get_empty_positions(board):
        board[i][j] = "O" #places the next move to check
        score = artificial_intelligence(board, 0, False) #calls the AI method, checks the next user input for wins
        board[i][j] = " "
        if score > best_score:
            best_score = score
            best_move = (i, j)
    board[best_move[0]][best_move[1]] = "O" #places the best move

def tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]
    print("Welcome to Tic Tac Toe!")
    user_first = input("Do you want to go first? (y/n): ").lower() == 'y'

    for _ in range(9):
        print_board(board)
        if user_first:
            user_move(board)
            if check_winner(board, "X"):
                print_board(board)
                print("Congratulations! You win!")
                return
            user_first = False
        else:
            computer_move(board)
            if check_winner(board, "O"):
                print_board(board)
                print("Computer wins! Better luck next time.")
                return
            user_first = True

        if not get_empty_positions(board):
            print_board(board)
            print("It's a draw!")
            return

if __name__ == "__main__":
    tic_tac_toe()
```

### Notes
Here is the TicTacToe Artificial intelligence implementation using the minmax function as the artificial intelligence. We can prove that the system works as working through the recurssion, the bestscore is the sum of all the moves that lead to a win, while alternating the user and computer. If there is a win, then the bestscore for those moves will be selected for. otherwise if there is no win condition possible, then the computer will play the first availibe move, as shown by the placement in the top left corner and subsequently top row if nothing is placed there.
